```python
# 1.Create a table rentals_may to store the data from rental table with information for the month of May.
engine.execute('create table rentals_may select*from rental where month(rental_date) = 5 and month(return_date) = 5')
```

```python
pd.read_sql('SELECT* FROM rentals_may', engine)
```

```python
# 2. Create a table rentals_june to store the data from rental table with information for the month of June.
engine.execute('create table rentals_june select*from rental where month(rental_date) = 6 and month(return_date) = 6')
```

```python
pd.read_sql('select*from rentals_june',engine)
```

```python
# 3. Check the number of rentals for each customer for May.
pd.read_sql('select count(rental_id), customer_id from rentals_may group by customer_id',engine)
```

```python
pd.read_sql('select count(rental_id), customer_id from rentals_june group by customer_id',engine)
```

```python
may=pd.read_sql('select count(rental_id) as rental_id, customer_id from rentals_may group by customer_id',engine)
june=pd.read_sql('select count(rental_id) as rental_id, customer_id from rentals_june group by customer_id',engine)
```

```python
rental_amount=may.merge(june,left_on='rental_id',right_on='rental_id',suffixes=('_may', '_june'))
```
