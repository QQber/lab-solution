```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('lab-customer-analysis-round-6/files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
data.head()
```

```python
# 1. Train-test split
x=data.drop('Total Claim Amount',axis=1)
y=data['Total Claim Amount']
```

```python
x=x.drop(columns=['Customer','Effective To Date'])
```

```python
x_num=x._get_numeric_data()
```

```python
x_cat=x.drop(columns=x_num.columns)
```

```python
print(x.shape,x_num.shape,x_cat.shape)
```

```python
x_num.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_scaled=StandardScaler().fit_transform(x_num)
```

```python
def normalize(x):
    x_mean=x.mean(axis=0)
    x_std=x.std(axis=0)
    x_std[x_std==0]=1
    x=(x-x_mean)/x_std
    return x
```

```python
x_num=normalize(x_num)
```

```python
x_num.head()
```

```python
(x_num-x_scaled).sum().sum()
```

```python
pd.get_dummies(x_cat,drop_first=True).head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
pd.DataFrame(OneHotEncoder(drop='first').fit_transform(x_cat).toarray(),columns=OneHotEncoder(drop='first').fit(x_cat).get_feature_names(input_features=x_cat.columns)).head()
```

```python
x_cat=pd.get_dummies(x_cat,drop_first=True)
```

```python
x=pd.concat([x_num,x_cat],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=.2)
```

```python
import warnings
warnings.simplefilter('ignore')
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
x_train_wi=sm.add_constant(x_train)
```

```python
model=sm.OLS(y_train,x_train_wi).fit()
```

```python
print(model.summary())
```

```python
# 2. Apply linear regression
from sklearn.linear_model import LinearRegression as LinReg
```

```python
linreg=LinReg()
```

```python
linreg.fit(x_train,y_train)
```

```python
y_pred_linreg=linreg.predict(x_test)
```

```python
y_pred_linreg
```

```python
# R2
print ('train R2: {} -- test R2: {}'.format(linreg.score(x_train, y_train),
                                            linreg.score(x_test, y_test)))
```

```python
# MSE
from sklearn.metrics import mean_squared_error as mse
```

```python
train_mse=mse(linreg.predict(x_train),y_train)
```

```python
test_mse=mse(linreg.predict(x_test),y_test)
```

```python
print('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
# RMSE
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
# MAE
from sklearn.metrics import mean_absolute_error as mae
```

```python
train_mae=mae(linreg.predict(x_train),y_train)
```
